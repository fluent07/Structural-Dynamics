```python
# Generate x values from 0 to 2π
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)

# Compute corresponding y values (sine function)
y = np.sin(x)
# Create a figure and axis
fig, ax = plt.subplots()

# Plot the sine wave
ax.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')

# Display the plot
plt.show()
```
